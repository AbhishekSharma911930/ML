```python
#3. defined inside another function
def pqr():
    def qrp():
        print("hello")
    qrp()
pqr()
```

```python
#4. return a func as ouput from another function
def pqr():
    def qrp():
        print("hello")
    qrp()
    return qrp
```

```python
#Anonymous function - Lambda Expression
# lambda inp_arg : output(returned)
# its always one liner , always returns , input arg can be multiple, optional, keyword, named etc.
#output can be simple/conditional/ comprehensionetc
  
```

```python
def doubler(x):
    return x*2
m=doubler(34)
print(m)
```

```python
# above function can be done as lambda expression as:
doubler_L = lambda x : x*2
doubler_L
```

```python
doubler_L(34)
```

```python
oddeve= lambda x: 'odd' if x%2 else 'even'
oddeve(45)
```

```python
# scope of variables
## global
## local
## nonloca
```

```python
x=10
def abc():
    print(x)
abc()
```

```python
x=10
def abc():
    x=30
    print(x)
print(x)
abc()
print(x)
```

```python
x=10
def abc():
    global x
    x=30
    #print(x)
print(x)
abc()
print(x)
```

```python
x=10
def abc():
    x=30
    def xyz():
        print('fromxyz',x)
    xyz()
print(x)
abc()
print(x)
```

```python
x=10
def abc():
    x=30
    global y
    y=15
    def xyz():
        global y
        y=28
        print('fromxyz',x)
    xyz()
    print(y)
print(x)
abc()
print(x)
```

```python
#above code can be done also as
x=10
def abc():
    x=30
    y=15
    def xyz():
        nonlocal y
        y=28
        print('fromxyz',x)
    xyz()
    print(y)
print(x)
abc()
print(x)
```
