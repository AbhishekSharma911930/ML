```python
from time import sleep
import cv2
import skimage as ski
import matplotlib.pyplot as plt
vid=cv2.VideoCapture(0)
flag, img=vid.read()
while True:
    flag, img=vid.read()
    if flag:
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        _, img_binary=cv2.threshold(img_gray,100,255,cv2.THRESH_BINARY)
        



        cv2.imshow('Preview',img_binary)
        key = cv2.waitKey(1)
        if key==ord('x'):
            break
    else:
        break
    sleep(0.1)
vid.release()
cv2.destroyAllWindows()
```

```python
from time import sleep
import cv2
import numpy as np
import skimage as ski
import matplotlib.pyplot as plt
fd= cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
vid=cv2.VideoCapture(0)
rand_color = tuple(np.random.random(size=3) * 256)
flag, img=vid.read()

while True:
    flag, img=vid.read()
    if flag:
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        x1,y1,w,h=(200,200,350,400)
        faces=fd.detectMultiScale(img_gray,1.1,5)
        for x1,y1,w,h in faces:
            #img_cropped=img[y1:y1+h, x1:x1+w, :1]
            
            cv2.rectangle(img, pt1=(x1,y1), pt2=(x1+w,y1+h),color=tuple(np.random.randint(size=3,low=0,high=255).tolist()),thickness=10)
        
        cv2.imshow('Preview',img)
        key = cv2.waitKey(1)
        if key==ord('x'):
            break
    else:
        break
    sleep(0.1)
vid.release()
cv2.destroyAllWindows()
```
