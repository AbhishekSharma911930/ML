```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1=cv2.imread('OIP.jpg')
img2=cv2.imread('OIP2.jpg')
img1=img1[50:50+262,0:50+474]
```

```python
img1.shape,img2.shape
```

```python
def showimg(im):
    plt.imshow(im[:,:,::-1])
    plt.show()
```

```python
def imshow(im):
    cv2.imshow('preview',im)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
imshow(img1)
imshow(img2)
```

```python
imshow(img2)
```

```python
added=cv2.add(img2,img1)
showimg(added)
```

```python
import numpy as np
```

```python
img2_1=cv2.multiply(img2,20*np.ones(img2.shape).astype('uint8'))
showimg(img2_1)
```

```python
img2_1=cv2.divide(img2,20*np.ones(img2.shape).astype('uint8'))
showimg(img2_1)
```

```python
##########FUNCTION FOR CHANGING CONTRAST using alpha and BRIGHTNESS using beta
```

```python
img2_2=cv2.convertScaleAbs(img2,alpha=1.3,beta=100)
showimg(img2_2)
```

```python
#################### Resizing the image
```

```python
img2_3=cv2.resize(img2, (400,400), interpolation=cv2.INTER_CUBIC)
img2_3.shape
```

```python
img2_4=cv2.resize(img2, None, fx=2, fy=0.5, interpolation=cv2.INTER_CUBIC)
img2_4.shape
```

```python
showimg(img2_4)
```

```python
import skimage as ski
```

```python
from skimage.transform import rotate
img2_5=rotate(img2,angle=60,resize=True)
showimg(img2_5)
```

# ANIMATION

```python
from time import sleep
repeat=True
while repeat:
    for i in range(0,361,5):
        i=rotate(img1,angle=i,resize=True)
        cv2.imshow('Preview',i)
        key=cv2.waitKey(1)
        if key==ord('x'):
            repeat=False
            break
        sleep(0.1)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
im_gray=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
im_r=img1[:,:,2].copy()
im_g = img1[:,:,1].copy()
im_b = img1[:,:,0].copy()
```
